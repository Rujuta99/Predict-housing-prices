# Feature Selection

```python
df = df.drop(["id","date","sqft_living","sqft_lot","sqft_above","sqft_living15","sqft_lot15"],axis=1)
df.head()
```

```python
fig = plt.figure(figsize=(7,5))
ax = fig.gca()
df['yr_renovated'].hist(grid=False, edgecolor='black')

plt.title('Histogram of renovated year')
plt.xlabel('Year')
plt.ylabel('No of houses')
```

```python
df.loc[df["yr_renovated"]>1,"yr_renovated"]=1
df['yr_renovated'].head(10)
```

# Model Building

```python
price = df['price']
y = np.array(price)
y
```

```python
x = df[['bedrooms','bathrooms','floors','waterfront','view','condition','grade','sqft_basement','yr_built','yr_renovated','zipcode','lat','long']]
x.head()
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.4,random_state=0)
```

```python
print('X Train',x_train.shape)
print('X Test',x_test.shape)
print('Y Train', y_train.shape)
print('Y Test', y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
```

# Model Evaluation

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_predict)*100
```

```python
fig = plt.figure(figsize=(7,5))
plt.scatter(y_test,y_predict)

plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
```
